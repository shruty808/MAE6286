```python
import numpy
from matplotlib import pyplot
%matplotlib inline
```

```python
# Set the font family and size to use for Matplotlib figures.
pyplot.rcParams['font.family'] = 'serif'
pyplot.rcParams['font.size'] = 16
```

```python
#set initial conditions:
c=1
nx=81
L=2.0
dx=L/(nx-1)
dt=0.02
nt=25
```

```python
x=numpy.linspace(0,L,num=nx)
```

```python
#Initial condition
u0=numpy.ones(nx)
r=numpy.where((0.5<=x)&(x<=1))
u0[r]=2
```

```python
u0
```

```python
pyplot.figure(figsize=(4.0,4.0))
pyplot.xlim(0.0,L)
pyplot.ylim(0,3.0)
pyplot.xlabel('x')
pyplot.ylabel('u0')
pyplot.title('Initial condition')
pyplot.grid()
pyplot.plot(x,u0,color='r',linestyle='--',linewidth=2)
```

```python
u=u0.copy()
for n in range(1,nt):
    un=u.copy()
    for i in range(1,nx):
        u[i]=un[i]-c*(dt/dx)*(un[i]-un[i-1])
    
```

```python
pyplot.plot(figsize=(4,4))
pyplot.grid()
pyplot.title('final ')
pyplot.xlim(0.0,L)
pyplot.ylim(0,3)
pyplot.plot(x,u0,color='r',linestyle='--',linewidth='2',label='inital')
pyplot.plot(x,u,color='b',linestyle='-',linewidth='2',label='nt=25')
pyplot.legend()
```
