```python
import numpy
from matplotlib import pyplot
%matplotlib inline
```

```python
nx=30
l=2.0
nt=10
sigma=0.2
dx=l/(nx-1)
v=0.3
dt=sigma*dx**2/v
x=numpy.linspace(0,l,num=nx)
```

```python
u0=numpy.ones(nx)
mask=numpy.where((x>0.5)&(x<1))
u0[mask]=2
```

```python
pyplot.plot(figsize=(4,4))
pyplot.xlim(0,l)
pyplot.ylim(0,3)
pyplot.title('initial condition')
pyplot.xlabel('x')
pyplot.ylabel('u0')
pyplot.grid()
pyplot.plot(x,u0,color='r',linestyle='--',linewidth=2)
```

```python
u=u0.copy()
for n in range(1,nt):
    un=u.copy()
    for i in range(1,nx-1):
        u[i]=u[i]+v*dt/(dx**2)*(u[i+1]+u[i-1]-2*u[i])
  
```

```python
pyplot.plot(figsize=(4,4))
pyplot.xlim(0,l)
pyplot.ylim(0,3)
pyplot.title('initial condition')
pyplot.xlabel('x')
pyplot.ylabel('u0')
pyplot.grid()
pyplot.plot(x,u,color='r',linestyle='--',linewidth=2)
pyplot.plot(x,u0,color='b',linestyle='--',linewidth=2)
```
