```python
import numpy
from matplotlib import pyplot
%matplotlib inline
```

```python
#initial conditions:
alpha=1.22e-3
L=1
nx=51
dx=L/(nx-1)
nt=100
x=numpy.linspace(0,L,num=nx)
T0=numpy.zeros(nx)
T0[0]=100
```

```python
def diff(t0,nt,dt,dx,alpha):
    t=T0.copy()
    for n in range (nt):
        t[1:-1]=t[1:-1]+alpha*(dt/dx**2)*(t[2:]-2*t[1:-1]+t[:-2])
    return t
```

```python
sigma=0.5
dt=sigma*dx**2/alpha
h=diff(T0,nt,dt,dx,alpha)
```

```python
#plot 
pyplot.figure(figsize=(5,5))
pyplot.xlabel('x')
pyplot.ylabel('temp')
pyplot.xlim(0,L)
pyplot.ylim(0,120)
pyplot.grid()
pyplot.plot(x,T0,color='r',linestyle='--',linewidth=2)
pyplot.plot(x,h,color='b',linestyle='-',linewidth=2)
```

```python
nt=1000
f=diff(T0,nt,dt,dx,alpha)
```

```python
pyplot.figure(figsize=(5,5))
pyplot.xlabel('x')
pyplot.ylabel('temp')
pyplot.xlim(0,L)
pyplot.ylim(0,120)
pyplot.grid()
pyplot.plot(x,T0,color='r',linestyle='--',linewidth=2)
pyplot.plot(x,f,color='b',linestyle='-',linewidth=2)
```

```python
def diff_with_bc(T0,nt,dt,dx,alpha):
    t=T0.copy()
    for n in range (nt):
        t[1:-1]=t[1:-1]+alpha*(dt/dx**2)*(t[2:]-2*t[1:-1]+t[:-2])
        t[-1]=t[-2]
    return t
```

```python
nt=2000
k=diff_with_bc(T0,nt,dt,dx,alpha)
```

```python
pyplot.figure(figsize=(5,5))
pyplot.xlabel('x')
pyplot.ylabel('temp')
pyplot.xlim(0,L)
pyplot.ylim(0,120)
pyplot.grid()
pyplot.plot(x,T0,color='r',linestyle='--',linewidth=2)
pyplot.plot(x,k,color='b',linestyle='-',linewidth=2)
```
