```python
import numpy
from matplotlib import pyplot
```

```python
def func1(nx,L=2.0,c=1.0,nt=20,sigma=0.5):
    dx=L/(nx-1)
    x=numpy.linspace(0,L,num=nx)
    dt=(sigma*dx)/c
    u0 = numpy.ones(nx)
    mask = numpy.where(numpy.logical_and(x >= 0.5, x <= 1.0))
    u0[mask] = 2.0
    u=u0.copy()
    for n in range(1,nt):
        u[1:]=u[1:]-c*(dt/dx)*(u[1:]-u[:-1])
    pyplot.plot(figsize=(4,4))
    pyplot.title('yay')
    pyplot.xlim(0,L)
    pyplot.ylim(0,4)
    pyplot.xlabel('x')
    pyplot.ylabel('u')
    pyplot.plot(x,u0,label='initial',color='r',linestyle='--',linewidth=2)
    pyplot.plot(x,u,label='final',color='b',linestyle='-',linewidth=2)
```

```python
func1(41)
```

```python
func1(81)
```
